```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('tagsets_json')
```

```python
text = """
Quantum physics is the fundamental study of matter and energy at the atomic and subatomic levels, revealing that nature operates in discrete packets called quanta rather than continuous flows. Developed in the early 20th century, this field challenges classical mechanics through principles like wave-particle duality, superposition, and entanglement, acting on both microscopic and macroscopic scales.
"""

# Convert to lowercase
text = text.lower()
print(text)
```

```python
words = word_tokenize(text)

print("Tokens:\n", words)
```

```python
pos_tags = pos_tag(words)

print("\nPOS Tags:\n", pos_tags)
```

```python
for word, tag in pos_tags:
    print(f"{word} --> {tag}")
```
