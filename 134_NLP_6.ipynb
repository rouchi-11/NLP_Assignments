```python
import nltk
import string
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report
from nltk.corpus import stopwords
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
texts = [
    "The team won the match",
    "The player scored a goal",
    "The match was exciting",
    "Government passed a law",
    "Election results were announced",
    "Political debate is ongoing",
    "New AI technology is emerging",
    "Machine learning is powerful",
    "Technology is evolving fast"
]

labels = [
    "Sports", "Sports", "Sports",
    "Politics", "Politics", "Politics",
    "Technology", "Technology", "Technology"
]
```

```python
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = nltk.word_tokenize(text)
    words = [w for w in words if w not in stop_words]
    return " ".join(words)

processed_texts = [preprocess(t) for t in texts]

print(processed_texts)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    processed_texts,
    labels,
    test_size=0.3,
    random_state=42,
    stratify=labels
)
```

```python
vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)

nb_pred = nb_model.predict(X_test_vec)
```

```python
svm_model = LinearSVC()
svm_model.fit(X_train_vec, y_train)

svm_pred = svm_model.predict(X_test_vec)
```

```python
print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print("\nNaive Bayes Report:\n", classification_report(y_test, nb_pred))

print("\nSVM Accuracy:", accuracy_score(y_test, svm_pred))
print("\nSVM Report:\n", classification_report(y_test, svm_pred))
```

```python
def predict_text(text, model):
    text = preprocess(text)
    vec = vectorizer.transform([text])
    return model.predict(vec)[0]

sample = "The match was exciting"

print("Naive Bayes Prediction:", predict_text(sample, nb_model))
print("SVM Prediction:", predict_text(sample, svm_model))
```
