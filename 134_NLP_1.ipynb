```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import wordnet
```

```python
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')
```

```python
text = """
Honeybees are highly social, beneficial insects belonging to the genus Apis, crucial for ecosystem health and agricultural pollination. They live in colonies consisting of one queen, thousands of female worker bees, and male drones, creating hives to store honey and rear young.
"""

# Convert to lowercase
text = text.lower()
print(text)
```

```python
words = word_tokenize(text)

print("Original Tokens:\n", words)
```

```python
ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in words]

print("\nStemmed Words:\n", stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

print("\nLemmatized Words:\n", lemmatized_words)
```

```python
print(text)
print("Original Tokens:\n", words)
print("\nStemmed Words:\n", stemmed_words)
print("\nLemmatized Words:\n", lemmatized_words)
```
