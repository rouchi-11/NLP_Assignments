```python
import nltk
import string
import numpy as np
!pip install gensim
from gensim.models import Word2Vec
```

```python
nltk.download('punkt_tab')
```

```python
text = """
Artificial intelligence is transforming the world. Artificial intelligence helps in automation.
Students are learning artificial intelligence.
"""

# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenization
tokens = nltk.word_tokenize(text)

print("Tokens:\n", tokens)
```

```python
# Word2Vec expects list of sentences
sentences = [tokens]
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,   # Embedding dimension (D)
    window=2,         # Context size
    min_count=1,
    sg=1,             # 1 = Skip-gram, 0 = CBOW
    epochs=100
)

print("Model trained successfully!")
```

```python
word = "intelligence"

vector = model.wv[word]

print(f"Embedding for '{word}':\n", vector)
```

```python
similar_words = model.wv.most_similar("artificial")

print("Words similar to 'artificial':")
for w, score in similar_words:
    print(w, "->", score)
```

```python
similarity = model.wv.similarity("artificial", "intelligence")

print("Similarity between 'artificial' and 'intelligence':", similarity)
```
